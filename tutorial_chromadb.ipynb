```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about New York",
        "This document is about Bangladesh"
    ],
    ids=['id1', 'id2']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents = collection.get(ids=["id1"])
documents
```

```python
results = collection.query(
    query_texts=['Query is about Hilsha Fish'],
    n_results=2
)
results
```

```python
collection.delete(ids=all_docs["ids"])
collection.get()
```

```python
collection.add(
    documents=[
        "This document is about New York",
        "This document is about Bangladesh"
    ],
    ids=["id1", "id2"],
    metadatas=[
        {"url":"https://en.wikipedia.org/wiki/New_York_City"},
        {"url":"https://en.wikipedia.org/wiki/Bangladesh"}
    ]          
)
```

```python
results = collection.query(
    query_texts=['Query is about Manhattan'],
    n_results=2
)
results
```
